```python
# Project topic:
#     You have to design a Food Ordering app for a restaurant.
# The application will have a log-in for admin and users.
# Admin will have the following functionalities:
# 1. Add new food items. Food Item will have the following details:
# 2. FoodID //It should be generated automatically by the application.
# 3. Name
# 4. Quantity. For eg, 100ml, 250gm, 4pieces etc
# 5. Price
# 6. Discount
# 7. Stock. Amount left in stock in the restaurant.
# 8. Edit food items using FoodID.
# 9. View the list of all food items.
# 10. Remove a food item from the menu using FoodID.

# The user will have the following functionalities:
# 1. Register on the application. Following to be entered for registration:
# 2. Full Name
# 3. Phone Number
# 4. Email
# 5. Address
# 6. Password
# 7. Log in to the application
# 8. The user will see 3 options:
# 9. Place New Order
# 10. Order History
# 11. Update Profile
# 12. Place New Order: The user can place a new order at the restaurant.
# 13. Show list of food. The list item should as follows:
# 14. 1 Tandoori Chicken (4 pieces) [INR 240]
# 15. 2 Vegan Burger (1 Piece) [INR 320]
# 16. 3 Truffle Cake (500gm) [INR 900]
# 17. Users should be able to select food by entering an array of numbers. For example, if the user wants to order Vegan 
#     Burger and Truffle Cake they should enter [2, 3]
# 18. Once the items are selected user should see the list of all the items selected. 
#     The user will also get an option to place an order.
# 19. Order History should show a list of all the previous orders
# 20. Update Profile: the user should be able to update their profile.
```

```python
import json
import sys
class food_app:
    def __init__(self):
        self.items={}
        self.food_id=len(self.items)+1
        self.admins={}
        self.admins_id=len(self.admins)+1
        self.users={}
        self.users_id=len(self.users)+1
        self.l=[]
        self.food_items={"1": {"Name": "Tandoori Chicken", "Quantity": "4 pieces", "Price": 240},"2": {"Name": "Vegan Burger", "Quantity": "1 Piece", "Price": 320},"3": {"Name": "Truffle Cake", "Quantity": "500gm", "Price": 900}}
#     Admin SignUp and login    
    def admin_registration(self):
        self.Name= input("Hello Admin, enter your full name: ")
        self.Ph_number= int(input("Enter your 10 digit mobile Number: "))
        self.Email= input("Enter your email address: ")
        self.Address= input("Enter your location address: ")
        self.Password= input("Enter alfanumeric password of 8 digit for your account: ")
        self.admin_data= {"Name":self.Name,"Mobile number":self.Ph_number,"Email":self.Email,"Address":self.Address,"Password":self.Password}
        self.admins_id=len(self.admins)+1
        self.admins[self.admins_id]=self.admin_data
        with open("admin_details.json","w+") as f:
            json.dump(self.admins,f)
        print("*******************************************************************************************")
        print(self.admins)
        print("Your are registred successfully")
    def admin_login(self):
        attempt=0
        print("You have only 3 attempts to login")
        while attempt<3:
            adminname=input("Enter your registered email id as adminname: ")
            password= input("Enter your registered password for your account: ")
            with open("admin_details.json","r") as f:
                file=json.load(f)
                for m in file:
                    for n in file[m]:
                        if n=="Email":
                            if adminname==file[m][n]:
                                with open("admin_details.json","r") as f:
                                    file=json.load(f)
                                    for m in file:
                                        for n in file[m]:
                                            if n=="Password":
                                                if password==file[m][n]:
                                                    print("You have successfully logged in")
                                                    return
                                                else:
                                                    attempt=attempt+1
                                                    print("incorrect password entered try again")
                            else:
                                attempt=attempt+1
                                print("incorrect email entered try again")
#     Admin functionalities                            
    def add_food(self):
        self.Name= input("Enter name of food: ")
        self.Quantity= int(input("Enter Quantity of food in KG you want to register: "))
        self.Price= int(input("Enter Price of food in Rupees for one plate: "))
        self.Discount= int(input("Enter total discount available in Rupees for one plate: "))
        self.Stock= int(input("Enter total unit of stocks available: "))
        self.food= {"Name":self.Name,"Quantity":self.Quantity,"Price":self.Price,"Discount":self.Discount,"Stock":self.Stock}
        self.food_id=len(self.items)+1
        self.items[self.food_id]=self.food
        with open("food_items.json","w+") as f:
            json.dump(self.items,f)
        print("*******************************************************************************************")
        print(self.items)
        print("Your food is added successfully")
    def delete_food_item(self):
        with open("food_items.json","r") as f:
            self.items=json.load(f)
            print(self.items)
            del self.items[input("Enter the food id you wnat to remove: ")]
            with open("food_items.json","w+") as f:
                json.dump(self.items,f)
                print("*******************************************************************************************")
                print(self.items)
                print("Your food is removed successfully")
    def Edit_food_items(self):
        with open("food_items.json","r") as f:
            self.items=json.load(f)
            print(self.items)
            edit_food_id= input("Enter the food id you want to edit: ")
            for i in self.items[edit_food_id]:
                if i=="Name":
                    self.items[edit_food_id][i]= input(f"Enter the {i} you want to update: ")
                else:
                    self.items[edit_food_id][i]= int(input(f"Enter the {i} you want to update: "))
        with open("food_items.json","w+") as f:
            json.dump(self.items,f)
        print("*******************************************************************************************")
        print(self.items)
        print("Your food is updated successfully")
    def view_food_items(self):
        with open("food_items.json","r") as f:
            self.items=json.load(f)
            print("MENU FOR FOOD")
            for i in self.items:
                print("*********************")
                print("Food id is",i,"\n*********************")
                for k in self.items[i]:
                    print("Food",k,"is",self.items[i][k])
                    print("*******************************************************************************************")
# user signUp and login
    def user_registration(self):
        self.Name= input("Hello, enter your full name: ")
        self.Ph_number= int(input("Enter your 10 digit mobile Number: "))
        self.Email= input("Enter your email address: ")
        self.Address= input("Enter your location address: ")
        self.Password= input("Enter alfanumeric password of 8 digit for your account: ")
        self.user_data= {"Name":self.Name,"Mobile number":self.Ph_number,"Email":self.Email,"Address":self.Address,"Password":self.Password}
        self.users_id=len(self.users)+1
        self.users[self.users_id]=self.user_data
        with open("user_details.json","w+") as f:
            json.dump(self.users,f)
        print("*********************************************************************************************************")
        print(self.users)
        print("Your are registred successfully")
        print("*********************************************************************************************************")
    def user_login(self):
        attempt=0
        print("You have only 3 attempts to login")
        print("*********************************************************************************************************")
        while attempt<3:
            username=input("Enter your registered email id as username: ")
            password= input("Enter your registered password for your account: ")
            print("*********************************************************************************************************")
            with open("user_details.json","r") as f:
                file=json.load(f)
                for m in file:
                    for n in file[m]:
                        if n=="Email":
                            if username==file[m][n]:
                                with open("user_details.json","r") as f:
                                    file=json.load(f)
                                    for m in file:
                                        for n in file[m]:
                                            if n=="Password":
                                                if password==file[m][n]:
                                                    print("You have successfully logged in")
                                                    return
                                                else:
                                                    attempt=attempt+1
                                                    print("incorrect password entered try again")
                                                    print("*********************************************************************************************************")
                            else:
                                attempt=attempt+1
                                print("incorrect email entered try again")
                                print("*********************************************************************************************************")
# user functionalities
    def update_user_details(self):
        update_users_id= int(input("Enter the user id you want to update: "))
        for i in self.users[self.users_id]:
            if i=="Ph_number":
                self.users[self.users_id][i]= int(input(f"Enter the {i} you want to update: "))
            else:
                self.users[self.users_id][i]= input(f"Enter the {i} you want to update: ")
        with open("user_details.json","w+") as f:
            json.dump(self.users,f)
        print("*********************************************************************************************************")
        print(self.users)
        print("*********************************************************************************************************")
        print("Your food is updated successfully")
        print("*********************************************************************************************************")
    def placing_new_order(self):
        print("MENU FOR FOOD")
        for i in self.food_items:
            print("*********************")
            print("Food id is",i,"\n*********************")
            for k in self.food_items[i]:
                print("Food",k,"is",self.food_items[i][k])
                print("*******************************************************************************************")
        order= (input("Enter the food id for placing order: "))
        for i in self.food_items:
            if i== order:
                print("Your order is placed for",self.food_items[i])
                self.l.append(self.food_items[i])
                break
        else:
            print("Enter food id is invalid try again")
    def order_history(self):
        print("Your order history is")
        print("*********************************************************************************************************")
        print(self.l)

c=food_app()
print("Enter what do want to do \n1 for signUp \n2 for login")
print("*********************************************************************************************************")
choose= int(input("Enter your choice: "))
print("*********************************************************************************************************")
if choose==1:
    print("Enter for your signUp as Admin or User or Exit \n1 for Admin \n2 for user \n3 for Exit")
    print("*********************************************************************************************************")
    x=int(input("Enter your choice for SignUp: "))
    print("*********************************************************************************************************")
    if x==1:
        c.admin_registration()
        print("*********************************************************************************************************")
    elif x==2:
        c.user_registration()
        print("*********************************************************************************************************")
    elif x==3:
        print("*********************************************************************************************************")
        sys.exit()
    else:
        print("It is not a valid input try again")
        print("*********************************************************************************************************")
elif choose==2:
    print("Enter for your login as Admin or User or Exit \n1 for Admin \n2 for user \n3 for Exit")
    print("*********************************************************************************************************")
    x=int(input("Enter your choice for login: "))
    print("*********************************************************************************************************")
    if x==1:
        c.admin_login()
        print("*********************************************************************************************************")
        print("You are logged in as Admin \nNow what do you want to perform \n1 for add food \n2 for Edit food items \n3 for delete food item \n4 for view food items")
        print("*********************************************************************************************************")
        y=int(input("Enter your choice for operation: "))
        print("*********************************************************************************************************")
        if y==1:
            c.add_food()
            print("*********************************************************************************************************")
        elif y==2:
            c.Edit_food_items()
            print("*********************************************************************************************************")
        elif y==3:
            c.delete_food_item()
            print("*********************************************************************************************************")
        elif y==4:
            c.view_food_items()
            print("*********************************************************************************************************")
        else:
            print("It is not a valid input try again")
            print("*********************************************************************************************************")
    elif x==2:
        c.user_login()
        print("*********************************************************************************************************")
        print("You are logged in as User \nNow what do you want to perform \n1 for Place New Order \n2 for Order History \n3 for Update Profile")
        print("*********************************************************************************************************")
        y=int(input("Enter your choice for operation: "))
        print("*********************************************************************************************************")
        if y==1:
            c.placing_new_order()
            print("*********************************************************************************************************")
        elif y==2:
            c.order_history()
            print("*********************************************************************************************************")
        elif y==3:
            c.update_user_details()
            print("*********************************************************************************************************")
        else:
            print("It is not a valid input try again")
            print("*********************************************************************************************************")
    elif x==3:
        sys.exit()
    else:
        print("It is not a valid input try again")
        print("*********************************************************************************************************")
```
